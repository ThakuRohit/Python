## Marker

# Different types of markers are :

```python
## Draw a 'Dashed' line having color 'Magenta' and point should be represented as 'diamond'.

import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([1,2,3,4])
y_points = np.array([5,2,9,4])

plt.plot(x_points, y_points, 'D--m')
plt.show()
```

```python
## Plot the line in graph having yellow color and points should be represented as "full plus"
## Note: Use "ms" to increase the size of marker by '20' and color the marker as black using 'mfc' and 'mec'

import matplotlib
from matplotlib import pyplot as plt
import numpy as np

x_points = np.array([1,2,3,4])
y_points = np.array([7,2,5,1])

plt.plot(x_points, y_points, 'y-P', ms=15, mfc='k', mec='k')
plt.plot()
```

```python
# PLot the points having two different points in x1, x2, y1, y2.

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 8, 4])

plt.plot(x1, y1, c='m')
plt.plot(x2, y2, lw=5)
plt.show()
```

## LABELS FOR A PLOT

```python
## Plotting the sport watch data of heartbeat and calories burnt. Also assign the label and title

import matplotlib.pyplot as plt
import numpy as np

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x,y, 'm:')
plt.title('Sport Watch Data', loc='left', fontdict = font1)
plt.xlabel('Average Pulse', fontdict = font2)
plt.ylabel('Calories Burned', fontdict = font2)
plt.show()
```

## Grid lines in plot

## Ploting in multiple graphs

```python
# Draw 2 plots on top of each other.

import matplotlib.pyplot as plt
import numpy as np

points_1 = np.array([1,2,3,4])
points_2 = np.array([4,7,1,8])

# plot-1
plt.subplot(2,1,1)
plt.plot(points_1, points_2)

# plot-2
plt.subplot(2,1,2)
plt.plot(points_2, points_1)

plt.show()
```

```python
# Add title in whole plot and also give title to each plot.

import matplotlib
from matplotlib import pyplot as plt
import numpy as np

# plot-1 
x = np.array([1,2,3,4])
y = np.array([10, 20, 30, 40])

plt.subplot(2,3,2)
plt.plot(x, y)
plt.title('Plot 1')

# plot-2
x = np.array([1,2,3,4])
y = np.array([3, 8, 1, 10])

plt.subplot(2,3,1)
plt.plot(x, y)
plt.title('Plot 2')

# plot-3
x = np.array([1,2,3,4])
y = np.array([10, 30, 20, 40])

plt.subplot(2,3,4)
plt.plot(x, y)
plt.title('Plot 3')

# plot-4
x = np.array([1,2,3,4])
y = np.array([7, 2, 9, 3])

plt.subplot(2,3,3)
plt.plot(x, y)
plt.title('Plot 4')

# plot-5
x = np.array([1,2,3,4])
y = np.array([40, 10, 30, 20])

plt.subplot(2,3,5)
plt.plot(x, y)
plt.title('Plot 5')

# plot-6
x = np.array([1,2,3,4])
y = np.array([2,7,3,9])

plt.subplot(2,3,6)
plt.plot(x, y)
plt.title('Plot 6')

# main 
plt.suptitle('Main')
plt.show()
```

## Scatter Plot

```python
# Example

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)
plt.xlabel('How Old the car is')
plt.ylabel('Car Speed')
plt.show()
```

```python
## Adding another scatter plot in same graph, assign color "hotpink" and "green" respectively using "c" or color.

import matplotlib.pyplot as plt
import numpy as np
 
# Plot-1
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y, c = "hotpink")

# Plot-2
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y, c='green')

plt.show()
```

```python
## We can assign different colors in each point using "c".

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x,y,c=colors)
plt.show()
```

## ColorMap or viridis 

```python
# Example

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x,y,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
## Example

import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s= sizes, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.show()
```

## BAR PLOT

```python
# Draw the bar vertically, assign color of bar as "yellow" with width as "0.9" along with grid lines. 

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y, color='y', width=0.9)
plt.grid()
plt.show()
```

```python
# Draw horizontal bar graph , height as 0.7 and give "hotpink" color to bar.

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.7, color='hotpink')
plt.show()
```

## PIE CHARTS

```python
# Example

import matplotlib.pyplot as plt
import numpy as np

value = np.array([25,35,10,30])

plt.pie(value)
plt.show()
```

```python
## Add label in each section using label

import matplotlib.pyplot as plt
import numpy as np

values = np.array([25,15,30,20])
label = np.array(['Gemini','Bard','ChatGPT','Alice'])
plt.pie(values, labels=label)
plt.show()
```

```python
# Change the angle of previous pie chart by 45 degree

import matplotlib.pyplot as plt
import numpy as np

values = np.array([25,15,30,20])
label = np.array(['Gemini','Bard','ChatGPT','Alice'])
plt.pie(values, labels=label, startangle=45)
plt.show()
```

```python
## Using "explode" parameter we can separate each section from pie chart

## Example: Separate "alice" chatbot section from the pie chart.

import matplotlib.pyplot as plt
import numpy as np

values = np.array([25,15,30,20])
label = np.array(['Gemini','Bard','ChatGPT','Alice'])
space = np.array([0,0,0,0.1])

plt.pie(values, labels = label, explode=space)
plt.show()
```

```python
## We can give 3-D effect, enabling "shadow" parameter "TRUE".

import matplotlib.pyplot as plt
import numpy as np

values = np.array([30,25,30,15])
label = np.array(['Chrome','Safari','Edge','Brave'])
space = np.array([0.2,0,0,0])

plt.pie(values, labels = label, explode=space, shadow = True)
plt.show()
```

```python
## Use Legend function to showcase each section shades separately.
## Note: Legend() can only be used when labels are added to chart

import matplotlib.pyplot as plt
import numpy as np

values = np.array([30,25,30,15])
label = np.array(['Chrome','Safari','Edge','Brave'])
space = np.array([0.2,0,0,0])

plt.pie(values, labels = label, explode=space, shadow = True)
plt.legend()
plt.show()
```

```python
## Adding Heading to the legend column

import matplotlib.pyplot as plt
import numpy as np

values = np.array([30,25,30,15])
label = np.array(['Chrome','Safari','Edge','Brave'])
space = np.array([0,0,0.2,0])

plt.pie(values, labels = label, explode=space, shadow = True)
plt.legend(title='Web Browser:')
plt.show()
```
